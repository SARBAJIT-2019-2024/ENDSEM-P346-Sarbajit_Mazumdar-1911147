```python
from My_Lib import *
import math
import matplotlib.pyplot as plt
import numpy as np

#making lists for appending the data as given
A=[]
X=[] 
Y=[]
with open("esem_fit1.dat") as C:
    for k in C:
        A.append(list(map(float, k.split())))
for i in range(len(A)):
    X.append(float(A[i][0]))  #Appending datas for X as temperature 
    Y.append(float(A[i][-1])) #Appending datas for Y as conductivity



for i in range(len(Y)):
    Y[i]=math.log(Y[i],math.e)

a,b,c,d,e,f=LeastSquare(X,Y)
print("Equation 1")
print("The value of (sigma_0) is ",math.exp(a)," and the slope alpha is ",b)
print("Pearson's r is calculated as: ",f)

x = np.arange(0, 9,0.2)
y=b*x+a
plt.plot(X, Y,'b.',label="data")
plt.plot(x,y,'--r',label="fitted curve")


plt.xlabel("$T(Temperature)$")
plt.ylabel("$log(\sigma)$")
plt.legend()
plt.show()

```
